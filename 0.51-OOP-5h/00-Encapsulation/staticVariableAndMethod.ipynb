```python
class testGetter:

    def __init__(self, name: str, age: int):
        self.__name = name
        self.__age = age


    @property
    def age(self):
        return self.__age
    
    @age.setter
    def age(self, value):
        if value <= 0:
            raise ValueError("Age can not be Negative...")
        self.__age = value

    
```
