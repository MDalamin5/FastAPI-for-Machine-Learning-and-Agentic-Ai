## Parent class

```python
## parent class
class User:
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender

    def login(self):
        print("Login Done...")

    def register(self):
        print("Register Done...")
```

## **Child class: Student**

```python
class Student(User):

    def __init__(self, name, gender, roll_no):
        super().__init__(name, gender)
        self.roll_no = roll_no

    def enroll(self):
        print(f"{self.name} enroll the course...")
```

## **Child class: Instructor**

```python
class Instructor(User):
    def __init__(self, name, gender, course_name):
        super().__init__(name, gender)
        self.course_name = course_name

    def publish_course(self):
        print(f"{self.name} publish the {self.course_name}...")
```

```python
stu1 = Student("Md Al Amin", "Male", 1001)
stu1.login()
```

```python
ins1 = Instructor("Aminul Islam", "Male", "CSE497")
```

```python
ins1.register()
ins1.login()
ins1.publish_course()
```
