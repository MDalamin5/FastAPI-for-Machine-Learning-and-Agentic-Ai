## **Types of Inheritance**
* Single Inheritance
* Multilevel Inheritance
* Hierarchical Inheritance
* Multiple Inheritance

## **Multilevel Inheritance**

```python
class GrandFather:
    pass

class Father(GrandFather):
    pass

class Child(Father):
    pass
```

## **Multiple Inheritance**

```python
class Father:
    def __init__(self):
        pass
    pass

class Mother:
    def __init__(self):
        pass
    pass

class child(Father, Mother):
    def __init__(self):
        super().__init__()
```

## **Heroical Inheritance**

```python
class Boss:
    pass

class Emp1(Boss):
    pass

class Emp2(Boss):
    pass
```
