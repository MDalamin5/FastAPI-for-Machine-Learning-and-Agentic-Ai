## **Polymorphism**
* Method Overriding
* Method Overloading
* Operator Overloading

```python
class Shape:
    def area(self,radius):
        return 3.14 * radius
    
    def area(self, h, b):
        return 0.5*h*b
    

s = Shape()
s.area(4)
```

## **In `python` do nor support Method Overloading**

```python
class Shape:
    def __init__(self):
        pass
    
    def area(self, radius=0, h=0, b=0):
        if radius == 0 and h != 0 and b != 0:
            return 0.5 * h * b
        
        if radius != 0 and h == 0 and b == 0:
            return 0.5 * radius
    

s = Shape()
s.area(4)
s.area(h=2, b=5)
```
